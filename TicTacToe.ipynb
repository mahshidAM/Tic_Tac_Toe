```python
from PIL import Image
from IPython.display import clear_output

bcolors_blue = '\033[94m'
bcolors_red = '\033[31m'
bold='\033[01m'
reset='\033[0m'
colors = {' ':' ','X':f'{bold}{bcolors_blue}X{reset}','O':f'{bold}{bcolors_red}O{reset}'}

# We make the board using dictonary in which keys be location (exemple: top-right ...) and initially with empty space as values
# Each move from player's choice of move will change the empty space values
theBoard = {'7': ' ' , '8': ' ' , '9': ' ' ,
            '4': ' ' , '5': ' ' , '6': ' ' ,
            '1': ' ' , '2': ' ' , '3': ' ' }

board_keys = [i for i in theBoard.keys()]

#Ask user if he/she wants to know the rules.
def rules():
    game_rules = input("Do you want to know the rules?(Y/N)")#display the sentence and ask to answer
    im = Image.open("img/tiktaktoe.png")

    if game_rules == "Y" or game_rules == "y":
        clear_output()
        print(f"\nThe game is played on a grid that's 3 squares by 3 squares.\nYou are "+colors['X']+", your friend is "+colors['O']+". Players take turns putting their marks in empty squares.\nThe first player to get 3 of her marks in a row (up, down, across, or diagonally) is the winner.\nWhen all 9 squares are full, the game is over. If no player has 3 marks in a row, the game ends in a tie. ")
        display(im)

    if game_rules == "N" or game_rules == "n":
        print("Oh excuse me Sir Madam i know everythings") 

#After every move in the game the updated borad need to be printed. We need to make a function in which we define the printBoard function. it'will permit to print board everytime by calling this function
def printBoard(board):   
    print("\n")
    print("\t     |     |")
    print(f"\t  "+colors[board['7']]+"  |  "+colors[board['8']]+"  |  "+colors[board['9']])
    print('\t_____|_____|_____')
 
    print("\t     |     |")
    print(f"\t  "+colors[board['4']]+"  |  "+colors[board['5']]+"  |  "+colors[board['6']])
    print('\t_____|_____|_____')
 
    print("\t     |     |")
 
    print(f"\t  "+colors[board['1']]+"  |  "+colors[board['2']]+"  |  "+colors[board['3']])
    print("\t     |     |")
    print("\n")

    
# Now we are going to make the main function with all gameplay functionnality.
def game():

    turn = 'X'
    count = 0
    
    im2 = Image.open("img/winloose.png")   

    while count < 10:
        printBoard(theBoard)
        print("It's your turn," + turn + ".In which place you want to move?") 

        try:
            move = input()
            if move not in board_keys:
                raise TypeError('please for the love of logic, choose a number include between 1 and 9 it\'s a tik tak toe')
            
            if theBoard[move] != ' ':# if the place where you have move is empty
                raise Exception('That place is already filled.\nIn which place you want to move?')
            else : #if have already a value, we display that's filled and you need to choose another place
                theBoard[move] = turn
                count += 1# the new value X (move) is stored in boardkey list                 
        except(TypeError,Exception) as error: 
            print(error)
            continue#repeating the operation

        # We need to check victory condition, if it's X or O has won for every move after 5 moves (because combination of 3 right move permit to win)
        if count >= 5:
            winner = f'*** {bold}{bcolors_blue}X{reset}  won. ****' if turn == 'X' else f'*** {bold}{bcolors_red}O{reset}  won. ****'

            if theBoard['7'] == theBoard['8'] == theBoard['9'] != ' ': 
                printBoard(theBoard)
                print("\nGame Over.\n")                
                print(winner)
                display(im2)
                break
            elif theBoard['4'] == theBoard['5'] == theBoard['6'] != ' ': 
                printBoard(theBoard)
                print("\nGame Over.\n")                
                print(winner)
                display(im2)
                break
            elif theBoard['1'] == theBoard['2'] == theBoard['3'] != ' ': 
                printBoard(theBoard)
                print("\nGame Over.\n")                
                print(winner)
                display(im2)
                break
            elif theBoard['1'] == theBoard['4'] == theBoard['7'] != ' ': 
                printBoard(theBoard)
                print("\nGame Over.\n")                
                print(winner)
                display(im2)
                break
            elif theBoard['2'] == theBoard['5'] == theBoard['8'] != ' ': 
                printBoard(theBoard)
                print("\nGame Over.\n")                
                print(winner)
                display(im2)
                break
            elif theBoard['3'] == theBoard['6'] == theBoard['9'] != ' ': 
                printBoard(theBoard)
                print("\nGame Over.\n")                
                print(winner)
                display(im2)
                break 
            elif theBoard['7'] == theBoard['5'] == theBoard['3'] != ' ': 
                printBoard(theBoard)
                print("\nGame Over.\n")                
                print(winner)
                display(im2)
                break
            elif theBoard['1'] == theBoard['5'] == theBoard['9'] != ' ': 
                printBoard(theBoard)
                print("\nGame Over.\n")                
                print(winner)
                display(im2)
                break 

        #A few smart people know they won't win but don't want to lose and try to obtain a Tie !
        #If neither players (X or O) wins and the board is full, it's a tie.
        if count == 9:
            print("\nGame Over.\n")                
            print("It's a Tie!!")
            break

        #We have to change the player after each move
        if turn =='X': #player 1
            turn = 'O' #switch to player 2
        else:
            turn = 'X' # if it's not player 1 but player 2 change to player 1           
    
    # we can ask if player wants to play again or not
    restart = input("Do want to play Again?(y/n)")
    if restart == "y" or restart == "Y":  
        for key in board_keys:
            theBoard[key] = " "

        game()

if __name__ == "__main__":
    rules()
    game()
```
